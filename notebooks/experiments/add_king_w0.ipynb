```python
# load table
from astropy.table import QTable

data = "../../data/mw_gc_parameters_orbital_structural_time.ecsv"
table = QTable.read(data, format="ascii.ecsv")
```

```python
import numpy as np

def king_concentration_parameter(core_radius, tidal_radius):
    return np.log10(tidal_radius / core_radius)
```

```python
# apply to the table
table["King_W"] = king_concentration_parameter(table["rc"], table["rt"])

# overwrite the table with the new column
table.write(data, format="ascii.ecsv", overwrite=True)
```
